```python
def calculate_tile_cost(width: float, height: float, cost_per_unit_area: float) -> float:
    
    area = width * height
    return area * cost_per_unit_area


def read_positive_number(prompt: str) -> float:
    
    while True:
        user_input = input(prompt).strip()
        try:
            value = float(user_input)
        except ValueError:
            print("Please enter a valid number.")
            continue

        if value < 0:
            print("Please enter a non-negative number.")
            continue

        return value


def main() -> None:
    print("Tile Cost Calculator")
    width = read_positive_number("Enter floor width: ")
    height = read_positive_number("Enter floor height: ")
    cost_per_unit_area = read_positive_number("Enter tile cost per unit area: ")

    total_cost = calculate_tile_cost(width, height, cost_per_unit_area)
    print(f"Total tile cost: {total_cost:.2f}")


if __name__ == "__main__":
    main()
```
